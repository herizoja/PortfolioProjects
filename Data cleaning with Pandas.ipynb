#  Cleaning Data using Pandas

```python
# import pandas 

import pandas as pd
```

```python
# import dataframe

df = pd.read_excel(r'/Users/herizojoseandre/Downloads/Customer Call List.xlsx')

df
```

```python
df.shape
```

```python
# drop the duplicate with the drop_duplicates() function

df = df.drop_duplicates()

df
```

```python
# remove not useful column by using the drop() function for the 'Not_Useful_Column'

df = df.drop(columns = 'Not_Useful_Column')

df
```

```python
# clean the column 'Last_Name' using the strip() function which takes from the left or right side

# df['Last_Name'] = df['Last_Name'].str.lstrip(' ')
# df['Last_Name'] = df['Last_Name'].str.lstrip('/')
# df['Last_Name'] = df['Last_Name'].str.lstrip('...')
# df['Last_Name'] = df['Last_Name'].str.rstrip('_')


# or another method using strip() function

df['Last_Name'] = df['Last_Name'].str.strip('123._/')

df
```

```python
# clean the column 'Phone_Number' using the replace() function

# df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z1-9]','')

# now format the column 'Phone_Number' using apply() function

# df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[3:10])

# df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))

# df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

# remove 'NaN' and 'N/a' from the column 'Phone_Number' using the function replace()

# df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')

df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')

df
```

```python
# Split the column 'Address' using the function split()

df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', 2, expand = True)

df
```

```python
# for the column 'Paying Cunstomer', replace the values 'Yes' to 'Y' and 'No' to 'N'

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
```

```python
# for the column 'Paying Cunstomer', replace the values 'N/a' to '' 

df['Paying Customer'] = df['Paying Customer'].str.replace('N/a', '')
```

```python
# for the column 'Dot_Not_Contact', replace the values 'Yes' to 'Y' and 'No' to 'N'

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df
```

```python
# now clean all 'N/a' values

df = df.replace('N/a', '')
```

```python
df
```

```python
# fill empty 'NaN' values

df = df.fillna('')
```

```python
df
```

```python
'''
Now, let's image the call center wants a list of the individuals they could call
'''

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
        
df
```

```python
# same method for the column 'Phone_Number'

for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
        
df

# another way to drop null values

# df = df.dropna((subset = 'Phone_Number'), inplace=True)
```

```python
# now reset the index

df = df.reset_index(drop=True)

df
```
